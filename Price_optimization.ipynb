## Step 2 : Data preparation

Steps:
- convert yes and nos to 0 & 1
- 3 category variable will be in furnishing
- recaling
- splitting data


> ### Steps
- Converting binary categorical variables to 0 and 1
- Convert multiple category variabes to dummy variables



```python
# yes no variables

varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
housing[varlist] = housing[varlist].apply(lambda x: x.map({'yes':1 , 'no': 0}))
housing.head()
```

```python
# creating dummy variables
varlist = ['furnishingstatus']
housing[varlist]
status  = pd.get_dummies(housing[varlist], drop_first= True)

```

```python
#concat dummy with original datafram

housing = pd.concat([housing, status] , axis=1)
housing.head()
```

```python

housing = housing.drop('furnishingstatus', axis=1)
housing.info()
```

Splitting data

```python
df_train, df_test = train_test_split(housing, train_size= 0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

Rescaling - imp when doing multiple regression

- Standard Scaling (Standardization)
- Min Max Scaling (Normalization)

Here we will use MinMax Scaling
- Min Max takes care of outliers

-> My note : Normalization group is more compressed than standardized data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#instantiate object
scaler = MinMaxScaler()
#fit on data
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
## fit () - computes x_max x_min etc
## tarnsform() - uses the x_max x_min

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

## Step 3: Training the model

How to arrive at final model?
- start with 13.remove one vaiable at a time
- start with 1. add 1 after the other and see model improvement

Here we start with a simple model and add one variable at a time manually

```python
#heatmap
fig = plt.figure(figsize = (16,12))
sns.heatmap(df_train.corr(), annot = True, cmap='YlGnBu')
plt.show()
```

We start with area as it is most corrleated

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
# adding variable and measuring statistical significance of variable
# variables which are corrleated with each other check VIF

import statsmodels.api as sm
```

```python
# make model with area
X_train_sm = sm.add_constant(X_train['area'])
```

```python
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
# significance of model

lr_model.summary()
```

Significance :

- p value is 0 so the result is statistically significant
- area only explains 28% of the variance of price

```python
#adding another variable : bathroom

X_train_sm = X_train[['area','bathrooms']]
X_train_sm = sm.add_constant(X_train_sm)
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
lr_model.summary()
```

Significance
- result is statistically significant
- 48% can be explained with area and bathroom

```python
#adding another variable : bathroom

X_train_sm = X_train[['area','bathrooms','bedrooms']]
X_train_sm = sm.add_constant(X_train_sm)
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
lr_model.summary()
```

Significance
- p value for constate rose a little but still result is statistically significant
- It explains 50% of the data. 2-3% jump is a good result

## Backward selection

### Adding all variables

```python
housing.columns
```

```python
var_list = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus_semi-furnished',
       'furnishingstatus_unfurnished']
X_train = df_train[var_list]

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
lr_model.summary()
```

We will make the inference of variable selection 
- p value
- VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### We drop semifurnished as p value is high

```python
X_train = X_train.drop('furnishingstatus_semi-furnished', axis=1)
```

```python
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### We remove bedroom as VIF is very high (VIF>5)

```python
X_train = X_train.drop('bedrooms', axis=1)
```

```python
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

print(lr_model.params)
```

```python
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Since all VIF are in optimal range, we finalize the parameters for our model

###  Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm)
res = y_train - y_train_pred
plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
#((y_train - y_train_pred), bins = 20)
plt.title("Residuals")

```

## Testing and evaluating the model

Steps :
- Scale the test set
- Predict

```python
# Scaling the test data set

#instantiate object

#fit on data
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
## fit () - computes x_max x_min etc
## tarnsform() - uses the x_max x_min

df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

```python
df_test.describe()
```

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_test_sm = sm.add_constant(X_test)
X_test_sm = X_test_sm.drop(["bedrooms", "furnishingstatus_semi-furnished"], axis = 1)
X_test_sm.info()
```

```python
# Make predictions

y_test_pred = lr_model.predict(X_test_sm)
```

```python
# Evaluate
from sklearn.metrics import r2_score
```

```python
r2_score(y_true = y_test , y_pred = y_test_pred)
```

### This r2 score is similar to that of the training set. Hence we can say that the model has generalized well.
